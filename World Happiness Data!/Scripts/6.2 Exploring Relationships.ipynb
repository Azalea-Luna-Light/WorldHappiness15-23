## 3. Exploring relationships

```python
# Creates a correlation matrix using pandas
df_hap.corr()
```

```python
# we need to drop the object columns as the correlation only functions with numeric values
df_sub = df_hap = df_hap.drop(columns = ['Happiness Rank', 'Year','Country', 'Region'])
```

```python
# Add labels, a legend, and change the size of the heatmap

f = plt.figure(figsize=(10, 10)) # figure size 
plt.matshow(df_sub.corr(), fignum=f.number) # type of plot
plt.xticks(range(df_sub.shape[1]), df_sub.columns, fontsize=10, rotation=90) # x axis labels
plt.yticks(range(df_sub.shape[1]), df_sub.columns, fontsize=10) # y axis labels
cb = plt.colorbar() # add a colour legend (called colorbar)
cb.ax.tick_params(labelsize=16) # add font size
plt.title('Correlation Matrix', fontsize=14) # add title
```

## Correlation Heatmap

```python
# Create a subplot with matplotlib
f,ax = plt.subplots(figsize=(9,9))

# Create the correlation heatmap in seaborn by applying a heatmap onto the correlation matrix and the subplots defined above.
corr = sns.heatmap(df_sub.corr(), annot = True, ax = ax) # The `annot` argument allows the plot to 
#place the correlation coefficients onto the heatmap.
```

### Discuss what the coefficients in the plot mean in terms of the relationships between the variables.
The greatest correlation is between the Healthy life Expectation factor and the factor Logged GDP per capita. This means that the assessment of Healthy life Expectation increases significantly as a result of the assessment of Ecomony (GDP per capita). 

The factors Logged GDP per capita and Healthy Life Expectancy seems to have the highest influence on the Happiness Score with 0.79 and 0,75- This means that the higher the GDP per capita and Healthy Life Expectancy , the higher the overall happiness score.

### 6. Create a categorical plot and interpret the results.

```python
# Highest Happiness score by region

# 1. set figure size
fig = plt.figure(figsize=(17,10) )

# 2. group by the region and the hapiness score
region_happiness_score = df.groupby('Region')['Happiness Score'].mean().sort_values().reset_index(drop = False)

# 3. Plot the graph
sns.barplot(x ='Region', y = 'Happiness Score', data = region_happiness_score,
                 palette = 'crest')
plt.title("Average Happiness score by region")
plt.xticks(rotation = 75)
```

```python
# fig = plt.figure(), # saving the image in the working directory 
fig.savefig("out.png") 
```

```python
df.loc[df['Happiness Score'] < 4.5, 'Happiness Score category'] = 'Low Happiness Score'
```

```python
df.loc[(df['Happiness Score'] >= 4.5) & (df['Happiness Score'] < 6), 'Happiness Score category'] = 'Middle Happiness Score'
```

```python
df.loc[df['Happiness Score'] >= 6, 'Happiness Score category'] = 'High Happiness Score'
```

```python
df['Happiness Score category'].value_counts(dropna = False)
```

```python
df.groupby('Region').agg({'Happiness Score': ['mean', 'min', 'max']})
```

```python
# Create a categorical plot in seaborn using the Happiness Score categories created above

sns.set(style="ticks")
g = sns.catplot(x="Region", y="Happiness Score", hue="Happiness Score category", data=df)
for axes in g.axes.flat:_= axes.set_xticklabels(axes.get_xticklabels(), rotation=90)
```

```python
# fig = g.get_figure(), # saving the image in the working directory 
g.savefig("out.png") 
```

### Oberservation
We can see that the regions North America and ANZ have only High Happiness Scores. However, they don't have the highest happiness scores. 

The highest happiness scores are found in Western Europe. 

And the region with the lowest Happiness Scores is South Asia and Middles East and North Africa.
